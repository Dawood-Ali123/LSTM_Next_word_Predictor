```python
from keras_preprocessing.text import Tokenizer 
tokenizer=Tokenizer()
```

```python
doc=doc.split('\n')
```

```python
tokenizer.fit_on_texts(doc)
```

```python
tokenizer.word_counts
```

```python
tokenizer.word_index #each value get a unique no 
```

```python
tokenizer.word_docs
```

```python
tokenized_sentence=tokenizer.texts_to_sequences(doc)
```

```python
tokenized_sentence
```

```python
input_sequences = []
for seq in tokenized_sentence:
    for i in range(1, len(seq)):
        input_sequences.append(seq[:i+1])
```

```python
input_sequences
```

```python
max_len=max([len(x) for x in input_sequences])
```

```python
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences 
paded_input=pad_sequences(input_sequences,maxlen=max_len,padding='pre' )
```

```python
 paded_input
```

```python
x=paded_input[:,:-1]
y=paded_input[:,-1 ]
```

```python
from keras.utils import to_categorical 
y=to_categorical(y,num_classes=377)
```

```python
from tensorflow.keras.models import Sequential 
from keras.layers import Embedding,Dense,LSTM 
```

```python
 model=Sequential()
model.add(Embedding(input_dim=377, output_dim=100, input_shape=(17,)))
model.add(LSTM(200))
model.add(Dense(377,activation='softmax'))
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.summary()
```

```python
 model.fit(x,y,epochs=100)
```

```python
text="I tried to debug my "
for i in range(10):
    #tokenize
    token_text=tokenizer.texts_to_sequences([text])[0]
    #padding 
    paded_token_text=pad_sequences([token_text],maxlen=17,padding='pre')
    #Model
    model.predict(paded_token_text)
    #prediction
    pos=np.argmax(model.predict(paded_token_text))
    for word,index in tokenizer.word_index.items():
        if index==pos:
            text=text+" "+word
            print(text)
```
